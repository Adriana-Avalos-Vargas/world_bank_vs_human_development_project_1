```python
import requests
import json
import pandas as pd
from IPython.display import clear_output
```

```python
#Lets retrieve form the API the following indicators for all the available countries.

#23806	Population with at least some secondary education (% ages 25 and older)
#44206	Population, total (millions)
#46006	Mobile phone subscriptions (per 100 people)
#52306	Research and development expenditure (% of GDP)
#53506	Foreign direct investment, net inflows (% of GDP)
#68606	GII: Gender Inequality Index, value
#128306	Prison population (per 100,000 people)
#137506	HDI: Human development index (HDIg) value
#137906	GDI: Gender Development Index
#185506	Population using at least basic drinking-water services (%)
#185606	Population using at least basic sanitation services (%)
#43606	Internet users (% of population)
#61006	Homicide rate (per 100,000 people)

#Lets make a list
indicators = [23806, 44206, 46006, 52306, 53506, 68606, 128306, 137506, 137906, 185506, 185606,43606, 61006]
total = len(indicators)
data_df = pd.DataFrame({'Country' : [], 'Year' : []})
count = 0

for kpi in indicators:
    url="http://ec2-54-174-131-205.compute-1.amazonaws.com/API/HDRO_API.php/indicator_id=" + str(kpi)
    
    response=requests.get(url).json()
    country_ID=[]
       
    for x in response["indicator_value"]:
        country_ID.append(x)
    
    year_dict=[]
    for x in country_ID:
        year_dict.append(response["indicator_value"][x][str(kpi)])
    
    temp_df=pd.DataFrame(year_dict)
    temp_df.index=country_ID
    temp_df.reset_index(inplace=True)
    temp_df.rename(columns={'index' : 'Country'}, inplace=True)
    temp_df = pd.melt(temp_df, id_vars = ["Country"], var_name="Year", value_name = kpi)
    temp_df = temp_df.set_index(['Country', 'Year'])
    
    if indicators.index(kpi) == 0:
        data_df = temp_df
    else:
        data_df = pd.merge(temp_df, data_df, on = ['Country', 'Year'])
        
    count += 1
    clear_output(wait=True)

    print(f'Iteration {count} of {total}')
    
data_df
```

```python
data_df.to_csv('human_development_data.csv')
```
