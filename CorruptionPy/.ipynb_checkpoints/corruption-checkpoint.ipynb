#### Importing Dependencies

```python
import pandas as pd
import numpy as np
```

##### Reading csv file from its path

```python
file = "CPI_DATA/Corruption_2019.csv"
cpi_df = pd.read_csv(file)
cpi_df.head()
```

#### Showing columns to rename and select the desired ones

```python
cpi_df.columns
```

#### Renaming desired columns

```python
cpi_df.rename(columns = {"Corruption Perceptions Index 2019: Global Scores " : "Country",
                         "Unnamed: 1" : "Country Code",
                         "Unnamed: 3" : "CPI", 
                         "Unnamed: 4" : "Rank"
}, inplace = True)

cpi_df.head()
```

#### Selecting only desired columns from the main df

```python
clean_df = cpi_df[["Country", "Country Code", "CPI", "Rank"]]
```

#### Droping the top 2 rows

```python
clean_df = clean_df.iloc[2:]
clean_df.head()
```

#### Resetting indeces

```python
final_df = clean_df.reset_index(drop=True)
final_df.head()
```

#### Writting a CSV file with the final data

```python
final_df.to_csv('CPI_DATA/CPI_final_data.csv')
```
